# KNN

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
def label_encode(df):
    df_labelled = df.copy()
    # Dictionary to input the different numbers for different classes
    label_encode = {"class": {"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2}}

    # Use .replace to change the different classes into numbers
    df_labelled.replace(label_encode,inplace=True)
    return df_labelled
```

```python
df = label_encode(df)
y_values = df['class']
```

```python
df = pd.read_csv("[Dataset]_Module_18_(iris).data",header=None)
names = ["sepal_length", "sepal_width","petal_length", "petal_width", "class"]
df.columns = names
df = label_encode(df)

x_values = df[['sepal_length','sepal_width','petal_length','petal_width']]
y_values = df['class']
```

# Test and train dataset (KNN)

```python
x_train, x_test, y_train, y_test = train_test_split(x_values,y_values,test_size=0.3,random_state=10)
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
#Train a K-Nearest Neighbours algorithm with the dataset. Remember to use the optimal number of neighbours (6) found in the earlier notebook.

KNN = KNeighborsClassifier(n_neighbors=6) #Sepertinya coba2
KNN = KNN.fit(x_train_scale,y_train)
```

```python
y_predict = KNN.predict(x_test_scale)
```

# Convolution matrix

```python
from sklearn.metrics import confusion_matrix  
print(confusion_matrix(y_test, y_predict))  
```

# Seaborn (Visualisasi conv matix)

```python
import seaborn as sns

# Create new dataframe
df = pd.DataFrame()

# Create new columns within dataframe
df['Actual'] = y_test
df['Predicted'] = y_predict

# Use pd.crosstab to count the frequency of the classes in actual and predicted
freq = pd.crosstab(df.Actual,df.Predicted)

# Use sns.heatmap to plot the heatmap
sns.heatmap(freq,annot=True,fmt="d")
```

# Decision Trees

```python
dt = tree.DecisionTreeClassifier(max_depth = 2, min_samples_split = 6)
dt.fit(x_train_scale,y_train)
```

# Convolutional matrix (decision trees)

```python
y_predict = dt.predict(x_test_scale)
```

```python
print(confusion_matrix(y_test, y_predict))  
```
